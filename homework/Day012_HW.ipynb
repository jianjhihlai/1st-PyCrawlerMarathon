### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
articles = {}


for d in soup.find(class_="part_list_2").find_all('h3'):
    tag = d.find('em').text
    if tag in articles:
        articles[tag].append(d.find_all('a')[-1].text)
    else:
        articles[tag] = [d.find_all('a')[-1].text]
        
print('ANS: 以dictionary儲存')
print(articles)
```

### ③ 哪一個類別的文章最多

```python

max_am = 0
tag_max = ''
for key, article in articles.items():
    n_articles = len(article)
    if n_articles > max_am:
        tag_max = key
        max_am = n_articles
    print('{}共有{}篇文章'.format(key, n_articles))

print('ANS: 最多文章的類別為{}'.format(tag_max))
```

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# options = webdriver.ChromeOptions()
# options.add_argument('--headless')
# options.add_argument('--no-sandbox')
# options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome(executable_path='./chromedriver/chromedriver.exe')
```

```python
import time
import datetime
today = datetime.datetime(2020, 1, 17)
news_time = datetime.datetime(2020, 1, 18)
browser.get("https://www.ettoday.net/news/news-list.htm")
def checkNewsTime(browser):
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    news = soup.find(class_="part_list_2").find_all('h3')[-1];
    check_time = datetime.datetime.strptime(news.find(class_="date").text, '%Y/%m/%d %H:%M')
    # print(check_time)
    return check_time
times_count = 0
print('今天時間：{}'.format(datetime.datetime.now()))
while news_time > today:
    time.sleep(5)
    browser.execute_script("window.scrollTo(0, 10000);")
    news_time = checkNewsTime(browser)
    times_count += 1
    if times_count > 300:
        break;

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = datetime.datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_time > today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
import os
import calendar
os.environ['TZ']='Asia/Taipei'

def current_time():
    d = datetime.datetime.utcnow()
    ts = calendar.timegm(d.timetuple())
    return datetime.datetime.fromtimestamp(ts).replace(microsecond=d.microsecond)

current = current_time()
before_2h = current - datetime.timedelta(hours=2)
print('現在時間：{}'.format(datetime.datetime.now()))
for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = datetime.datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_time > before_2h:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import os
import calendar
import datetime
import time
os.environ['TZ']='Asia/Taipei'
today = datetime.datetime(2020, 1, 18)
news_time = datetime.datetime(2020, 1, 19)
day_before_3d = today - datetime.timedelta(days=3)
start_time = day_before_3d.replace(hour=15)
end_time = day_before_3d.replace(hour=17)
browser.get("https://www.ettoday.net/news/news-list-2020-1-15-0.htm")
def checkNewsTime(browser):
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    news = soup.find(class_="part_list_2").find_all('h3')[-1];
    check_time = datetime.datetime.strptime(news.find(class_="date").text, '%Y/%m/%d %H:%M')
#     print(check_time)
    return check_time
times_count = 0
print('今天時間：{}'.format(datetime.datetime.now()))
while news_time > start_time:
    time.sleep(5)
    # browser.execute_script("window.scrollTo(0, 10000);")
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    # browser.send_keys(Keys.PAGE_DOWN)
    # browser.perform()
    news_time = checkNewsTime(browser)
    times_count += 1
    if times_count > 300:
        break;

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = datetime.datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_time < end_time:
        if news_time > start_time:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### 1.（簡答題）檔案、API、爬蟲三種取得資料方式有什麼不同？

AND: 


*   檔案：使用簡易，適合表格式資料。
*   API：適合複雜且需要控管的資料。
*   爬蟲：在沒有檔案或API提供資料來源時，僅能透過爬蟲收集資料，較無結構，處理較為麻煩。




### 2.（實作）完成一個程式，需滿足下列需求：
    * 下載指定檔案到 Data 資料夾，存成檔名 Homework.txt
    * 檢查 Data 資料夾是否有 Homework.txt 檔名之檔案
    * 將「Hello World」字串覆寫到 Homework.txt 檔案
    * 檢查 Homework.txt 檔案字數是否符合 Hello World 字數


```python
# 根據需求引入正確的 Library

from urllib.request import urlretrieve
import os
```

```python
# 下載檔案到 Data 資料夾，存成檔名 Homework.txt

try:
    url = 'https://www.w3.org/TR/PNG/iso_8859-1.txt'
    filename = './Data/Homework.txt'
    datafolder = './Data'
    os.makedirs(datafolder, exist_ok=True )
    urlretrieve(url, filename=filename)
except:
    print('發生錯誤！')
```

```python
# 檢查 Data 資料夾是否有 Homework.txt 檔名之檔案

files = []

files = [f for f in os.listdir(datafolder) if os.path.isfile(os.path.join(datafolder, f))]

if 'Homework.txt' in files:
    print('[O] 檢查 Data 資料夾有 Homework.txt 檔名之檔案')
else:
    print('[X] 檢查 Data 資料夾無 Homework.txt 檔名之檔案')
```

```python
# 將「Hello World」字串覆寫到 Homework.txt 檔案

f = ''

with open("./Data/Homework.txt", "w") as fh:
    fh.write("Hello World")

try:
    with open("./Data/Homework.txt", "r") as fh:
        f = fh.read()

except EnvironmentError: # parent of IOError, OSError *and* WindowsError where available
    pass    
    
```

```python
# 檢查 Homework.txt 檔案字數是否符合 Hello World 字數

if len('Hello World') == len(f):
    print('[O] 檢查 Homework.txt 檔案字數符合 Hello World 字數')
else:
    print('[X] 檢查 Homework.txt 檔案字數不符合 Hello World 字數')
```

### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異
#### ANS:
* r.text是字串型態，要存取其中的值需要用parser才行，例如json套件
* json.loads(r.text) 將r.text的字串parser成字典型態，可以用index和key做存取



```python
# r.text
import requests

r = requests.get('https://api.github.com/events')
r.text
```

```python
import json

a= json.loads(r.text)
```

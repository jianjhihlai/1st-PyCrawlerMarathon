```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://github.com')
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

```python
fp = open('github.txt', 'r')
credential = fp.read()
user, pw = credential.split(',')
fp.close()
```

```python
sess = requests.session()
payload = {
    'commit': 'Sign in',
    'utf8': '%E2%9C%93',
    'login': user,
    'password': pw,
}
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36',
}
res = sess.get('https://github.com/session', headers=headers)
soup = BeautifulSoup(res.text, 'html5lib')
payload['authenticity_token'] = soup.find('input', attrs={'name': 'authenticity_token'})['value']
res = sess.post('https://github.com/session', data=payload, headers=headers)
res = sess.get('https://github.com/new', headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

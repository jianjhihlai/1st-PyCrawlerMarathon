## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

````python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
````

````python
import requests
url = 'http://www.zhihu.com'

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'
}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
````

### 2. http://aicoin.cn

````python
import requests

url = 'https://www.aicoin.cn/'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
````

````python
from IPython.display import Image
Image(filename='aicoin_headers.png')
````

````python
import requests
url = 'https://www.aicoin.cn/'

headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'
}

r = requests.get(url, headers=headers)
print(r.text[0:600])
````

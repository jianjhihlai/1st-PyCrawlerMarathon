## 作業目標

* 根據範例提供的 API ，完成以下問題：

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

import requests
import json

headers = {'user-agent': 'my-app/0.0.1'}

# 取出問題
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text

data = json.loads(response)

```

```python
from datetime import datetime

questionTime = datetime.fromtimestamp(data['data'][0]['question']['created'])
print('55493026問題的發問時間：{:%Y-%m-%d %H:%m:%S}'.format(questionTime))
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

# number of total answers
totalAnswers = data['paging']['totals']
limit = len(data['data'])
print('回答總數：{:d}'.format(totalAnswers))

# get last page
offset = totalAnswers - limit
urlLast = 'https://www.zhihu.com/api/v4/questions/55493026/answers?limit='+str(limit)+'&offset='+str(offset)

r = requests.get(urlLast,headers=headers)
response = r.text

lastPage = json.loads(response)

countLastPage = len(lastPage['data'])
```

```python
firstAnswerTime = datetime.fromtimestamp(data['data'][0]['created_time'])
print('第一筆的回答時間：{:%Y-%m-%d %H:%m:%S}'.format(firstAnswerTime))

lastAnswerTime = datetime.fromtimestamp(lastPage['data'][countLastPage - 1]['created_time'])
print('最後一筆的回答時間：{:%Y-%m-%d %H:%m:%S}'.format(lastAnswerTime))
```

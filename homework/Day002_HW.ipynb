```python
import os
import urllib.request

res = "http://opendata.hccg.gov.tw/dataset/432257df-491f-4875-8b56-dd814aee5d7b/resource/de014c8b-9b75-4152-9fc6-f0d499cefbe4/download/20150305140446074.csv"
datafolder = './data'

os.makedirs(datafolder, exist_ok=True )

urllib.request.urlretrieve(res, './data/example.csv')
```

### CSV Reader

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

```python
# 1. 取出班次一的每一個時間
try:
    idx = 0
    timeOfBus = ''
    with open('./data/example.csv', newline='') as csvfile:
        # 讀取 CSV 檔案內容
        rows = csv.reader(csvfile)
        # 以迴圈輸出每一列
        for row in rows:
            if idx == 0:
                c = row.index('班次1')
                timeOfBus += row[c] + '：'
            else:
                timeOfBus += row[c] + ';'
            idx += 1
    print(timeOfBus)
except e:
    print('沒有班次1！')
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存
# 用時間串列儲存
import time as t
import csv

bus1_time = []
try:
    idx = 0
    with open('./data/example.csv', newline='') as csvfile:
        # 讀取 CSV 檔案內容
        rows = csv.reader(csvfile)
        # 以迴圈輸出每一列
        for row in rows:
            if idx == 0:
                c = row.index('班次1')
            else:
                bus1_time.append(t.strptime(row[c], '%H:%M'))
            idx += 1
    print(bus1_time)
except e:
    print('沒有班次1！')
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存
# 用dict儲存
import time as t
import csv

bus_idx = {}
bus_time = {}
try:
    idx = 0
    with open('./data/example.csv', newline='') as csvfile:
        # 讀取 CSV 檔案內容
        rows = csv.reader(csvfile)
        # 以迴圈輸出每一列
        for row in rows:
            if idx == 0:
                bus_idx = {val: idx for idx, val in enumerate(row) if val.startswith('班次')}
            else:
                for bus, c in bus_idx.items():
                    bus_time[bus] = t.strptime(row[c], '%H:%M')
            idx += 1
    print(bus_time)
except e:
    print('沒有班次1！')
```
